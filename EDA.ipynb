**INSIGHTS** :

2. Order table has missing values in some date columns. These records with missing values in some date columns make sense since these records have incompleted order status. 3. A small number of missing values in product table is identified in other columns beside "product_id" column. Missing values in category column will be filled by "Uncategorized"
4. Order review table lacks lots of record in review_comment_title and review_comment_message.
